```python
#Caminante aleatorio en 2D
import matplotlib.pyplot as plt 
import numpy as np
from scipy.spatial import distance

print("----- Caminante aleatorio 2D -----")
print("Este programa simula un número m de repeticiones de caminatas de n pasos.")
print("El caminante solo se puede mover arriba, abajo, a la derecha y a la izquierda con igual probabilidad.")
print("Introduzca el número de repeticiones:")
m = int(input()) # m = número de repeticiones
print("Introduzca el número de pasos que dará el caminante:")
n = int(input()) # n = numero de pasos que dará el caminante
print("----- Resultados -----")
print("Para  m=",m,"y n=",n)

# Comenzamos con definir la trayectoria del RW
distancias = np.zeros(m) #Vector donde se guardan las distancias
radios = np.zeros(m) #Vector donde se guardan los radios
plt.figure(figsize=(10,5)) #Tamaño de la trama (gráfica)

# Simulación 
for k in range(0,m): 
    x = np.zeros(n) #Vector de ceros de la componente x
    y = np.zeros(n) #Vector de ceros de la componente y
    for i in range(1,n):
        paso = np.random.randint(1, 5) #Elegimos un valor entero al azar de la siguinte lista [1,2,3,4]
        x[0]=0 # Todos las repeticiones comienzan en el origen
        y[0]=0
        punto_inicial = np.array([x[0],y[0]]) # Punto inicial del RW
        if paso == 1: #Si es 1, el caminante se mueve una unidad a la derecha en x y cero en y.
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1]
        elif paso == 2: #Si es 2, el caminante se mueve una unidad a la izquierda en x y cero e y.
              x[i] = x[i - 1] - 1
              y[i] = y[i - 1]
        elif paso == 3: #Si es 3, el caminante se mueve una unidad hacia arriba en y y cero en x.
              x[i] = x[i - 1]
              y[i] = y[i - 1] + 1
        else: #Si no es 1, 2 y 3 entonces es 4, se mueve una unidad hacia abajo en y y cero en x.
              x[i] = x[i - 1]
              y[i] = y[i - 1] - 1 
             
    punto_final = np.array([x[n-1],y[n-1]]) # Punto final del caminante después de n pasos        
    distancias[k] =  (distance.euclidean(punto_final,punto_inicial))**2 # Se calcula la distancia entre el punto final y el inicial de cada caminata
    suma=0
    for i in range(0,n):
        for j in range(0,n):
            suma = suma + (x[i]-x[j])**2 + (y[i]-y[j])**2
    Rg = (1/2*(n**2))*suma       
    radios[k] = Rg
    #Graficamos cada trayectoria sobre la misma trama
    #plt.title("Caminante aleatorio $2D$ (Número de repeticiones $m=" + str(m) + "$)  (Número de pasos $n=" + str(n) + "$)",fontsize=8)
    #plt.plot(x, y,"-",linewidth=1)
    #plt.xlabel("x")
    #plt.ylabel("y")
#plt.plot(x[0],y[0],"rx", markersize=13)       
#plt.grid()
#plt.show()

#Calculamos propiedades estadísticas
#Media de la distancia principio - fin
distancia_media = np.mean(distancias)
#Media del radio de giro
radio_medio = np.mean(radios)

print("El valor esperado de la distancia principio - fin:")
print("<(d(P_i,P_f))^2> =",distancia_media)

print("El valor esperado del radio de giro:")
print("<Rg^2> =",radio_medio)


```

```python
#Caminata aleatoria cuántica 
import math
import pylab


#Definimos la función de probabilidad
def probabilities(posn):
    """Devuelve una lista de probabilidades para cada lugar"""
    return [sum([abs(amp) ** 2 for amp in place]) for place in posn]


def normalise(posn):
    """Normalizar función para normalizar una línea de entrada 1D"""
    N = math.sqrt(sum(probabilities(posn)))
    return [[amp / N for amp in place] for place in posn]


def timestep(posn):
    """Define la acción de un paso de tiempo, es decir, una puerta Hadamard en cada elemento"""
    return normalise([[x[0] + x[1], x[0] - x[1]] for x in posn])


def shift(coin):
    """Mueva los elementos arriba hacia la izquierda y los elementos abajo hacia la derecha"""
    newposn = [[0, 0] for i in range(len(coin))]
    for j in range(1, len(posn) - 1):
        newposn[j + 1][0] += coin[j][0]
        newposn[j - 1][1] += coin[j][1]
    return normalise(newposn)


# Inicializamos la lista
min, max = -100,101
posn = [[0, 0] for i in range(min, max)]
posn[-min] = [1 / math.sqrt(2), 1j / math.sqrt(2)]

# Correr para algunos pasos
for time in range(-min):
    posn = shift(timestep(posn))

# Graficamos la distribuición de onda
pylab.plot(range(min, max), probabilities(posn),"r")
pylab.xlabel("Posición")
pylab.ylabel("Probabiliad de encontrar al CA en $x$")
pylab.title("Densidad de probabilidad del CA cuántico")
pylab.grid()
pylab.savefig("quantum_random.png",bbox_inches="tight",dpi=200)
pylab.show()
```

```python
import random
import matplotlib.pyplot as plt

prob = [0.5, 0.5]  # Probability to move up or down

start = 0  #Start at 2
positions = [start]

for _ in range(1, 100):
    rr = random.random()
    down = rr < prob[0] and positions[-1] > 1
    up = rr > prob[1] and positions[-1] < 4
    positions.append(positions[-1] - down + up)

plt.plot(positions)
plt.show()
```
