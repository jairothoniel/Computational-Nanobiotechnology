```python
#Aplicaciones de la nanotecnología computacional
#Proyecto 4 - Modelo de juguete para la evolución entre especies
#Jair Domínguez, Adriel Reyes, Eduardo Flores 
#Este codigo es un modelo de evolución biológica simple, se basa en generar un vector que contenga en cada componente
#el número de especiesnasignar una regla de interacción y dejar evolucionar.

#Importamos las librerias necesarias para la creción de matrices y graficas
import numpy as np
import matplotlib.pyplot as plt 

#Leemos los parámetros libres que da el usuario
print("Ingrese el número de especies que contendrá el vector:")
N=int(input())
print("Número de especies=",N)
print("Ingrese el número de pasos temporales de evolución:")
steps=int(input())
print("Número de pasos temporales=",steps)

#Creación de la figura para el estado inicial
fig, ax = plt.subplots()
ax = plt.axes(xlim=(1,N), ylim=(0,1))
plt.xlabel('Especies')
plt.ylabel('Número de especies normalizado')
plt.savefig("grafica_inicial.jpg")

#Creamos un arreglo con valores aleatorios entre [0,1] y de dimensión igual al número de especies
especies = np.random.uniform(0, 1, N) 

plt.plot(np.arange(1,N + 1), especies,'yo')
plt.bar(np.arange(1, N + 1), especies,color="violet")
plt.title('Estado inicial, promedio = {}'. format(np.mean(especies)))
plt.savefig("grafica_inicial2.jpg")
plt.show()


l = [1, steps/4, steps/2, steps] # Indicamos cada cuando se va a crear una gráfica

#Aplicamos el algoritmo de interacción
for i in range(2, steps + 1):
       j = np.random.randint(0,N-2) #Creamos un número aleatorio, especie que interacciona de forma aleatoria
       if especies[j] > especies[j+1]: #Condición de interacción
              #Regla de evolución 1
              #especies[j+1] = (especies[j-2] + especies[j+2])/3
              #Regla de evolución 2
              especies[j-1] = np.cos((especies[j-1] + especies[j+1]))/5
       if i in l:
              plt.title('Número de pasos evolutivos= {}, promedio = {}'. format(i, np.mean(especies)))
              plt.plot(np.arange(1,N + 1), especies, 'go')
              plt.bar(np.arange(1,N + 1), especies,color="magenta")
              plt.xlabel('Especies')
              plt.ylabel('Número de especies normalizado')
              plt.savefig("grafica_{}.jpg". format(i) )
              plt.show()
       
```
